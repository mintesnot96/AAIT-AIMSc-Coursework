```python
import matplotlib.pyplot as plt
threads = [1, 2, 4, 8, 16, 16]
sequential_times = [5.417573, 5.406174, 5.416678, 5.449090, 5.425162, 2.493525]
parallel_times = [5.433323, 3.012772, 1.764181, 1.260950, 1.234066, 0.673186]
speedups = [0.997101, 1.794418, 3.070364, 4.321416, 4.396168, 3.704065]

fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.plot(threads[:-1], speedups[:-1], 'o-', color='blue', label='Speedup')

for i, (thread, seq_time, par_time, spd) in enumerate(zip(threads[:-1], sequential_times[:-1], parallel_times[:-1], speedups[:-1])):
    ax1.annotate(f'Seq: {seq_time:.2f}s\nPar: {par_time:.2f}s', 
                 (thread, spd), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center')

ax1.scatter(threads[-1], speedups[-1], color='red', label='-O3 Optimization')
ax1.annotate(f'-O3 Seq: {sequential_times[-1]:.2f}s\nPar: {parallel_times[-1]:.2f}s', 
             (threads[-1], speedups[-1]), 
             textcoords="offset points", 
             xytext=(0,10), 
             ha='center')


ax1.set_xlabel('Number of Threads')
ax1.set_ylabel('Speedup', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
plt.title(' Speedup  vs. Number of Threads and Execution Time \nApproximation of  using Pthreads \n\n')

ax2 = ax1.twinx()


ax2.plot(threads[:-1], sequential_times[:-1], 's--', color='orange', label='Sequential Time')
ax2.plot(threads[:-1], parallel_times[:-1], '^--', color='green', label='Parallel Time')
ax2.set_ylabel('Execution Time (s)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')


ax1.grid(True)


plt.show()
```
