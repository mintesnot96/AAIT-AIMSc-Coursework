```python
import matplotlib.pyplot as plt

overhead_values = {
    '16x1': {
        1024: 17.37761944,
        731: 54.532032740000005,
        512: 60.85860511,
        365: 50.7286464,
        256: 63.672758280000004
    },
    '1x16': {
        1024: 24.93144375,
        731: 30.863198699999998,
        512: 41.441,
        365: 49.7318645,
        256: 64.67275828
    },
    '8x2': {
        1024: 37.48168921,
        731: 46.25557997,
        512: 53.30507861,
        365: 61.52941001,
        256: 70.12981209
    },
    '2x8': {
        1024: 27.992095189999997,
        731: 39.099857990000004,
        512: 45.47197168,
        365: 47.904481010000005,
        256: 52.04871586
    },
    '4x4': {
        1024: 34.5538321,
        731: 47.2395799,
        512: 58.335078610000004,
        365: 66.54941001,
        256: 71.832487209
    }
}

geometry_colors = {
    '16x1': 'orange',
    '1x16': 'blue',
    '8x2': 'green',
    '2x8': 'red',
    '4x4': 'yellow'
}

correct_N_order = [1024, 731, 512, 365, 256]

fig, ax = plt.subplots(figsize=(14, 8))

for geometry, data in overhead_values.items():
    overhead_list = [data[N] for N in correct_N_order]
    line, = ax.plot(correct_N_order, overhead_list, marker='o', label=geometry, linestyle='-', linewidth=2.5, color=geometry_colors[geometry], markersize=8)
    for N, overhead in zip(correct_N_order, overhead_list):
        ax.annotate(f"{overhead:.2f}%", 
                    (N, overhead),
                    textcoords="offset points",
                    xytext=(0,10),
                    ha='center',
                    fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=geometry_colors[geometry], lw=2))

ax.set_xlabel('Grid Size (N=N/1.4)', fontsize=14)
ax.set_ylabel('Communication Overhead (%)', fontsize=14)
ax.set_title('Communication Overhead for Different Geometries (2D)', fontsize=16)
ax.legend(title='Geometry', fontsize=12, title_fontsize='13', shadow=True, fancybox=True)
ax.set_xticks(correct_N_order)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_facecolor('white')
fig.patch.set_facecolor('white')
ax.grid(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the results
configs = [
    "Serial",
    "1MPI + 2OpenMP",
    "2MPI + 2OpenMP",
    "4MPI + 2OpenMP",
    "8MPI + 2OpenMP",
    "16MPI + 2OpenMP",
    "1MPI + 4OpenMP",
    "2MPI + 4OpenMP",
    "4MPI + 4OpenMP",
    "1MPI + 8OpenMP",
    "2MPI + 8OpenMP",
    "4MPI + 8OpenMP"
]

gflops_rates = [
    7.39912,
    7.32065,
    12.37072,
    21.84672,
    32.95702,
    31.51502,
    7.24072,
    12.53722,
    22.46242,
    7.38234,
    12.72672,
    20.69632
]
colors = ['gray',  
          'skyblue', 'skyblue',
          'skyblue', 'skyblue', 'skyblue', 
          'lightgreen', 'lightgreen', 'lightgreen', 
          'salmon', 'salmon', 'salmon'] 

plt.figure(figsize=(14, 8))
bars = plt.bar(configs, gflops_rates, color=colors)
plt.xlabel('Configuration', fontsize=14)
plt.ylabel('GFlops Rate', fontsize=14)
plt.title('MPI+OpenMP hybrid parallelism with 1D geometry', fontsize=16)
plt.xticks(rotation=45, ha='right')


from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='gray', label='Serial'),
                   Patch(facecolor='skyblue', label='2 OpenMP Threads'),
                   Patch(facecolor='lightgreen', label='4 OpenMP Threads'),
                   Patch(facecolor='salmon', label='8 OpenMP Threads')]


for bar, rate in zip(bars, gflops_rates):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(rate, 2), ha='center', va='bottom', fontsize=8)

plt.legend(handles=legend_elements, title='OpenMP Threads')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


mpi_openmp_configurations = [
    "1MPI +2OpenMP", "1MPI +8OpenMP", "1MPI +4OpenMP",
    "2MPI +8OpenMP", "2MPI +4OpenMP", "2MPI +2OpenMP",
    "4MPI +2OpenMP", "4MPI +4OpenMP", "4MPI +8OpenMP",
    "8MPI +2OpenMP", "16MPI +2OpenMP"
]
mpi_openmp_gflops = [
    7.32065, 7.38234, 7.24072,
    12.72672, 12.53722, 12.37072,
    21.84672, 22.46242, 20.69632,
    32.95702, 31.51502
]


mpi_only_gflops = [
    7.32075, 11.22853, 21.91571, 21.80054, 55.9912
]


mpi_only_gflops_mapped = []
for config in mpi_openmp_configurations:
    mpi_count = int(config.split("MPI")[0])
    if mpi_count == 1:
        mpi_only_gflops_mapped.append(mpi_only_gflops[0])
    elif mpi_count == 2:
        mpi_only_gflops_mapped.append(mpi_only_gflops[1])
    elif mpi_count == 4:
        mpi_only_gflops_mapped.append(mpi_only_gflops[2])
    elif mpi_count == 8:
        mpi_only_gflops_mapped.append(mpi_only_gflops[3])
    elif mpi_count == 16:
        mpi_only_gflops_mapped.append(mpi_only_gflops[4])


plt.figure(figsize=(14, 8))


plt.plot(mpi_openmp_configurations, mpi_openmp_gflops, label='MPI+OpenMP', marker='o', linestyle='-', color='darkorange', linewidth=2, markersize=8)
for i, rate in enumerate(mpi_openmp_gflops):
    plt.text(mpi_openmp_configurations[i], rate + 0.5, f"{rate:.2f}", fontsize=9, ha='center')


plt.plot(mpi_openmp_configurations, mpi_only_gflops_mapped, label='MPI Only', marker='x', linestyle='-', color='green', linewidth=2, markersize=8)
for i, rate in enumerate(mpi_only_gflops_mapped):
    plt.text(mpi_openmp_configurations[i], rate - 1.5, f"{rate:.2f}", fontsize=9, ha='center')

plt.xticks(rotation=45, ha="right", fontsize=12) 
plt.yticks(fontsize=12)
plt.ylabel('Gflops Rate', fontsize=14) 
plt.title('Performance Comparison: MPI Only vs Hybrid(MPI+OpenMP)', fontsize=16)  
plt.tight_layout()  
plt.show() 
```
